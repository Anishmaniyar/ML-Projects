```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Set up a 2x2 grid of separate plots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))

# 2. Plot each histogram in its designated grid position (ax)
sns.histplot(data=df, x='Tenure', bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Tenure')

sns.histplot(data=df, x='Usage Frequency', bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Usage Frequency')

sns.histplot(data=df, x='Support Calls', bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Support Calls')

sns.histplot(data=df, x='Payment Delay', bins=20, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Payment Delay')

sns.histplot(data=df, x='Total Spend', bins=20, kde=True, ax=axes[2, 1])
axes[2, 1].set_title('Distribution of Total Spend')

sns.histplot(data=df, x='Last Interaction', bins=20, kde=True, ax=axes[2, 2])
axes[2, 2].set_title('Distribution of Last Interaction')

# 3. Clean up the spacing so labels don't overlap
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Initialize a 2x2 grid of separate subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(14, 10))

# 2. Plot 1: Tenure vs Churn (Top Left)
sns.kdeplot(data=df, x='Tenure', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Tenure by Churn')

# 3. Plot 2: Usage Frequency vs Churn (Top Right)
sns.kdeplot(data=df, x='Usage Frequency', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Usage Frequency by Churn')

# 4. Plot 3: Support Calls vs Churn (Bottom Left)
sns.kdeplot(data=df, x='Support Calls', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Support Calls by Churn')

# 5. Plot 4: Payment Delay vs Churn (Bottom Right)
sns.kdeplot(data=df, x='Payment Delay', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Payment Delay by Churn')

sns.kdeplot(data=df, x='Total Spend', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5, ax=axes[2, 1])
axes[2, 1].set_title('Distribution of Total Spend by Churn')

sns.kdeplot(data=df, x='Last Interaction', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5, ax=axes[2, 2])
axes[2, 2].set_title('Distribution of Last Interaction by Churn')

# 6. Clean up spacing so layouts look neat
plt.tight_layout()
plt.show()
```

```python
#Correlation Matrix
corr_matrix = df.corr(numeric_only=True)

# 2. Plot it as a colorful heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# 1. Define the exact order of categories for ordinal columns
subscription_order = ['Standard', 'Premium']
contract_order = ['Monthly', 'Quarterly', 'Annually']

# 2. Set up the ColumnTransformer with both Ordinal and One-Hot encoders
preprocessor = ColumnTransformer(
    transformers=[
        ('ordinal_sub', OrdinalEncoder(categories=[subscription_order]), ['subscription type']),
        ('ordinal_con', OrdinalEncoder(categories=[contract_order]), ['contract length']),
        ('onehot_age', OneHotEncoder(sparse_output=False), ['Gender']) # Added One-Hot Encoding here
    ],
    remainder='passthrough'
)
```
